```python
import requests
```

```python
url = 'https://covid19-brazil-api.now.sh/api/report/v1/'
```

```python
response = requests.get(url)
response #output 200 = OK
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(response.json()['data'])
df
```

```python
df.info()
```

```python
df.isna().any()
```

```python
df.duplicated().any()
```

```python
df.nunique()
```

```python
df.drop(columns=['uid', 'uf', 'datetime', 'suspects', 'refuses'], inplace=True)
df
```

```python
df.columns = [
    'Estado',
    'Casos',
    'Obitos'
]
df
```

```python
df.to_csv('cleaned_covid.csv', index=False)
```

```python
df = pd.read_csv('cleaned_covid.csv')
df
```

```python
confirmados = df[['Estado','Casos']].sort_values(by='Casos', ascending=False) [:10]
confirmados
```

```python
import matplotlib.pyplot as plt
plt.barh(confirmados['Estado'], confirmados['Casos'])
plt.title('Top 10 Estados com mais Casos de COVID-19')
plt.xlabel('Casos')
plt.ylabel('Estado')
plt.show()
```

```python
obitos = df[['Estado','Obitos']].sort_values(by='Obitos', ascending=False) [:10]
obitos
```

```python
plt.barh(obitos['Estado'], obitos['Obitos'])
plt.title('Top 10 Estados com mais Óbitos por COVID-19')
plt.xlabel('Óbitos')
plt.ylabel('Estado')
plt.show()
```

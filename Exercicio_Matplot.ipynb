**1. Reproduza a imagem abaixo utilizando a biblioteca matplot.**

pie_demo_features1.webp

```python
import matplotlib.pyplot as plt

valor = [15,30,45,10]
rotulo = ['Frogs', 'Hogs', 'Dogs', 'Logs']
deslocar = [0,.1,0,0]

plt.figure(figsize = (5,5))

plt.pie(valor,
        explode=deslocar,
        labels=rotulo,
        autopct="%.1f%%",
        pctdistance=0.6,
        startangle=90,
        shadow = True
        )

plt.show()
```

**2. Reproduza a imagem abaixo utilizando a biblioteca matplot.**

scatter_plot_drinks_3.fe5d87ad562c.avif

```python
import matplotlib.pyplot as plt
import numpy as np
preco = np.asarray([2.50, 1.23, 4.02, 3.25, 5.00, 4.40])
vendas = np.asarray([34, 62, 49, 22, 13, 19])
tamanho = np.asarray([20, 35, 40, 20, 27.5, 15])
cores = np.asarray(['green', 'red', 'yellow', 'yellow', 'red', 'green'])
plt.figure(figsize=(12,10))
plt.scatter(x=preco, y=vendas, s=tamanho*20, c=cores)
plt.show()
```

**3. Reproduza a imagem abaixo utilizando a biblioteca matplot.**

```python
import matplotlib.pyplot as plt
linguagem = ['Python', 'C++','Java','Perl','Scala','Lisp']
ranking = [10,8,6,4,2,1]
plt.figure(figsize=(8,8))
plt.bar(x=linguagem,height=ranking,color='mediumslateblue', edgecolor='black')
plt.xlabel('Linguagem de Programao')
plt.ylabel('Ranking')
plt.title('Ranking de Linguagens de Programao')
plt.ylim(0,10)
plt.show()
```
